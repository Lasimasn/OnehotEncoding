```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('mercedes.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data
```

```python
#checking how many unique categories are in each variable 
for col in data.columns:
    print(col, ':',len(data[col].unique()),'labels')
```

```python
#finding top 10 most frequent categories for X2
data.X2.value_counts().head(20)
```

```python
#Making a list with the top 10 categories
top10=[x for x in data.X2.value_counts().head(10).index]
top10
```

```python
# one hot encoding for 10 variables
for label in top10:
    data[label]=np.where(data['X2']==label,1,0)
data[['X2']+top10].head(5)
```

```python
#Get whole set of dummy variables for all categorical variables 
def one_hot(df,variable,top_x_labels):
    for label in top_x_labels:
        df[variable+'_'+label]=np.where(data[variable]==label,1,0)
        
data=pd.read_csv('mercedes.csv',usecols=['X1','X2','X3','X4','X5','X6'])

one_hot(data,'X2',top10)
data.head()
```

```python
#one hot encoding on X1
X1_top10=[x for x in data.X1.value_counts().head(10).index]

one_hot(data,'X1',X1_top10)
data.head()
```

```python
#one hot encoding on X5
X5_top10=[x for x in data.X5.value_counts().head(10).index]

one_hot(data,'X5',X5_top10)
data.head(10)
```

```python
#one hot encoding on X6
X6_top10=[x for x in data.X6.value_counts().head(10).index]

one_hot(data,'X6',X6_top10)
data.head()
```
